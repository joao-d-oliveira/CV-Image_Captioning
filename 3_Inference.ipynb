```python
from model import EncoderCNN, DecoderRNNv101, DecoderRNNv102, DecoderRNNv110, DecoderRNNv120, DecoderRNNv121
import pickle
import torch
import os
from vocabulary import Vocabulary

def loadTrainedModels(run, epoch):
    encoder_file = run + f'encoder-{epoch}.pkl'
    decoder_file = run + f'decoder-{epoch}.pkl'
    batchsize = int(run[run.find('_batch_size') + 11:run.find('_vocabThr')])
    embed_size = int(run[run.find('embedSize') + 9:run.find('_hiddenSize')])
    hidden_size = int(run[run.find('_hiddenSize') + 11:run.find('_totEpochs')])
    smallDB = True if 'small' in run else False
    
    if smallDB: 
      voca = pickle.load(open('/content/vocab_small_training_thr4.pkl', 'rb'))
      vocab_size = 5892
    else: 
      voca = pickle.load(open('/content/vocab_complete_training_thr4.pkl', 'rb'))
      vocab_size = len(voca)

    encoder = EncoderCNN(embed_size)
    model = run[:run.find('_')]
    if model == 'v101' or model == 'v100':  decoder = DecoderRNNv101(embed_size, hidden_size, vocab_size)
    elif model == 'v102':  decoder = DecoderRNNv102(embed_size, hidden_size, vocab_size)
    elif model == 'v110':  decoder = DecoderRNNv110(embed_size, hidden_size, vocab_size)
    elif model == 'v120':  decoder = DecoderRNNv120(embed_size, hidden_size, vocab_size)
    elif model == 'v121':  decoder = DecoderRNNv121(embed_size, hidden_size, vocab_size)

    # Move models to GPU if CUDA is available. 
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    encoder.to(device)
    decoder.to(device)
    print(run)
    encoder.load_state_dict(torch.load(os.path.join('./models', encoder_file), map_location=device) )
    decoder.load_state_dict(torch.load(os.path.join('./models', decoder_file), map_location=device))

    return encoder, decoder

#encoder, decoder = loadTrainedModels(FILENAME, 3)

def clean_sentence(data_loader, output):
    sentence = [data_loader.dataset.vocab.idx2word[f] for f in output]
    if sentence[0] != data_loader.dataset.vocab.start_word or sentence[-1] != data_loader.dataset.vocab.end_word: 
      print('incorrect starting/ending sentence')
    #separate words by space, expect last char, given it should be a end of sentence '.', '!',...
    sentence=" ".join(sentence[1:-2]) + sentence[-2]
    return sentence
```

```python
# (Optional) TODO: Validate your model.

def generateValDict(encoder, decoder, vocab_file='vocab_complete_training_thr4.pkl'):
    import numpy as np
    import matplotlib.pyplot as plt
    from torchvision import transforms
    from data_loader import get_loader
    %matplotlib inline

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")    

    transform_test = transforms.Compose([ 
        transforms.Resize(256),                          # smaller edge of image resized to 256
        transforms.RandomCrop(224),                      # get 224x224 crop from random location
        #transforms.RandomHorizontalFlip(0.5),               # horizontally flip image with probability=0.5
        transforms.ToTensor(),                           # convert the PIL Image to a tensor
        transforms.Normalize((0.485, 0.456, 0.406),      # normalize image for pre-trained model
                            (0.229, 0.224, 0.225))])


    # Create the data loader.
    vocab_file = '/content/' + vocab_file
    data_loader_val = get_loader(transform=transform_test, vocab_file=vocab_file, 
                                 vocab_from_file=True ,mode='val', cocoapi_loc='')
    result = []
    for i, (idx, orig_image, image, caption) in enumerate(data_loader_val):
        print(f'\rReading {i:<5} of {len(data_loader_val)}', end='')
        image = image.to(device)
        features = encoder(image).unsqueeze(1)
        output = decoder.sample(features.squeeze(1))
        
        a = dict()
        a['image_id'] = idx
        a['caption'] = clean_sentence(data_loader_val, output)
        
        result.append(a)
    
    return result

results = generateValDict(encoder, decoder, 'vocab_small_training_thr4.pkl')
```
